```python
# Plot data using the plot() method of the axes object
# Simple XY plot
# Generate data for plots
x = [1, 2, 3, 4, 5]
y = x

# Get an empty figure
fig1 = plt.figure()

# Get the axes instance at 1st location in 1x1 grid
ax = fig1.add_subplot(1,1,1)

# Generate the plot
ax.plot(x, y)

# Set labels for x and y axis
ax.set_xlabel('X--->')
ax.set_ylabel('Y--->')

# Set title for the plot
ax.set_title('Simple XY plot')

# Display the figure
plt.show()
```

```python
# Multiple subplots in the same figure
# Function to get the square of each element in the list
def list_square(a_list):
    return [element**2 for element in a_list]

# Multiple subplots in same figure
fig3 = plt.figure()
x = [1, 2, 3, 4, 5]
y = x
z = list_square(x)

# Divide the figure into 1 row 2 column grid and get the
# axes object for the first column
ax1 = fig3.add_subplot(1,2,1)

# plot y = x on axes instance 1
ax1.plot(x, y)

# set x and y axis labels
ax1.set_xlabel('X------>')
ax1.set_ylabel('Y------>')
ax1.set_title('Y=x plot')

# Get second axes instance in the second column of the 1x2 grid
ax2 = fig3.add_subplot(1,2,2)

# plot z = x^2
ax2.plot(x, z)
ax2.set_xlabel('X------>')
ax2.set_ylabel('Z=x^2------>')
ax2.set_title('Z=x^2 plot')

# Generate the title for the Figure. Note that this is different then the title for individual plots
fig3.suptitle("Two plots in a figure")
plt.show()
```

**Seaborn:** Seaborn is a Data Visualization library that uses Matplotlib underneath to plot graphs. It is easier to plot complex graphs with less code as compared to Matplotlib. Let's learn it.

Seaborn has *built-in datasets* for us. One of the datasets is the Restaurant Customer Bill and Tips Dataset. Customer visit the rstaurant for Lunch or Dinner. Dataset records the amount of total bill along with the amount given to waiters as tip. Dataset also gives other details about customer such as sex smoker or not, day of the week, time either Lunch or dinner and size as number of people visited together. Let's import the dataset.

```python
# Let's Import Seaborn as sns and all other needed libaries
import seaborn as sns

# Tips Dataset has been imported as DataFrame
tips = sns.load_dataset('tips')
```
